<h1>Now Finding categorical columns </h1>

```python
categorical_values = []
for i in dset.columns:
    if dset[i].dtype == "object":
        categorical_values.append(i)
        
print("Categorical columns :", categorical_values)
```

```python
from sklearn import preprocessing
le= preprocessing.LabelEncoder()
```

```python
for i in categorical_values:
    dset[i] = le.fit_transform(dset[i])
```

```python
dset.head()
```

```python
X= dset.drop('Burned_transformers_2019',axis='columns')
Y= dset.Burned_transformers_2019
```

```python
#feature selection using stats Model
import statsmodels.api as sm
X= sm.add_constant(X) 
```

```python
lr = sm.OLS(Y,X).fit()
print(lr.summary2())
```

```python
p_values= lr.pvalues
cols= p_values[p_values<=0.05].index.tolist() 
print(cols)
```

```python
X= X[cols[1:]]
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, stratify= Y, random_state = 42)
```

```python
## checking for the distribution of traget variable in train test split
print('Distribution of target variable in training dataset')
print(Y_train.value_counts())

print('Distribution of target variable in test dataset')
print(Y_test.value_counts())
```

<h1> Data Normalization </h1>

```python
from sklearn.preprocessing import MinMaxScaler
sc= MinMaxScaler()
X_train= sc.fit_transform(X_train)
X_test= sc.transform(X_test)
```

<h1> Decision Tree </h1>

SK Learn_DT

```python
from sklearn.tree import DecisionTreeClassifier

# create an instance of the DecisionTreeClassifier class
clf= DecisionTreeClassifier(max_depth= 4,random_state= 42)
clf.fit(X_train,Y_train)
```

```python
predY= clf.predict(X_test)
"""### *Accuracy Score*"""

from sklearn.metrics import accuracy_score,f1_score
print("Model Accuracy: {0}%".format(accuracy_score(Y_test, predY)))
print("f1_score ",f1_score(Y_test, predY,average='weighted'))
```

XGBoost_DT

```python
from xgboost import XGBClassifier

# Create an instance of XGBClassifier
xgb_clf= XGBClassifier(max_depth= 4,random_state= 42)
xgb_clf.fit(X_train,Y_train)
```

```python
predY= xgb_clf.predict(X_test)
"""### *Accuracy Score*"""

from sklearn.metrics import accuracy_score,f1_score
print("Model Accuracy: {0}%".format(accuracy_score(Y_test, predY)))
print("f1_score ",f1_score(Y_test, predY,average='weighted'))
```

<h1> Random Forest </h1>

XG Boost Random Forest

```python
from xgboost import XGBRFClassifier
rfxb= XGBRFClassifier(n_estimators=100, max_depth=4)
rfxb.fit(X_train,Y_train)
```

```python
predY= rfxb.predict(X_test)
"""### *Accuracy Score*"""

from sklearn.metrics import accuracy_score,f1_score
print("Model Accuracy: {0}%".format(accuracy_score(Y_test, predY)))
print("f1_score ",f1_score(Y_test, predY,average='weighted'))
```
